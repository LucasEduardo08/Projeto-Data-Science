```python
from pickle import load
from sklearn.neural_network import MLPClassifier
```

```python
# Carregando os dados de treino
with open("titanic.pkl", "rb") as f:
    x_treino, x_teste, y_treino, y_teste = load(f)
```

```python
# Verificando as formas
x_treino.shape, y_treino.shape
```

```python
x_teste.shape, y_teste.shape
```

```python
# Fazendo o treinamento utilizando redes neurais

rede_neural_titanic = MLPClassifier(max_iter=1000, verbose=True)
rede_neural_titanic.fit(x_treino, y_treino)
```

```python
# Fazendo as previsões 
previsoes = rede_neural_titanic.predict(x_teste)
previsoes
```

```python
# fazendo as comparações
y_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Verificando a precisão do modelo
accuracy_score(y_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
# Analisando os resultados
cm = ConfusionMatrix(rede_neural_titanic)
cm.fit(x_treino, y_treino)
cm.score(x_teste, y_teste)
```

```python
print(classification_report(y_teste, previsoes))
```
